```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gradient_descent(x,y):
	m_curr = y_curr = 0
	iterations = 51
	n = len(x)
	learning_rate = 0.8

	plt.scatter(x,y,color='red',marker="o",linewidth=3)
	plt.xlabel('X')
	plt.ylabel('Y-Predicted')
	for i in range(iterations):
		y_pred = m_curr * x + y_curr
		plt.plot(x, y_pred, color='green')
		cost = (sum([val for val in ((y-y_pred)**2)]))/n
		md = -(2/len(x)) * sum(x*(y-y_pred))
		bd = -(2/len(x)) * sum(y-y_pred)

		m_curr = m_curr - learning_rate * md
		y_curr = y_curr - learning_rate * bd
		print("m {}, b {}, cost {}, iteration {}".format(m_curr,y_curr,cost,i))
```

```python
x = np.array([92,56,88,70,80,49,65,35,66,67])
y = np.array([98,68,81,80,83,52,66,30,68,73])
```

```python
gradient_descent(x,y)
```
