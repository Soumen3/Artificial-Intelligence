```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.svm import SVC
```

```python
df = pd.DataFrame(load_iris().data, columns=load_iris().feature_names)
```

```python
df.head()
```

```python
df['flower'] = load_iris().target
df['flower'] = df.flower.apply(lambda x: load_iris().target_names[x])
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop(['flower'], axis='columns'), df.flower, test_size=0.2)
```

```python
model = SVC(kernel='rbf', C=30, gamma='auto')
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
import numpy as np
```

```python
kernels = ['rbf', 'linear', 'sigmoid', 'poly']
C = [1, 10, 20, 30]
avg_scores = {}
for kval in kernels:
	for cval in C:
		cv_scores = cross_val_score(SVC(kernel=kval, C=cval, gamma='auto'), df.drop(['flower'], axis='columns'), df.flower, cv=5)
		avg_scores[kval + '_' + str(cval)] = np.average(cv_scores)

avg_scores
```

## The above function we can perform using the API GridSearchCV of sklearn.

```python
from sklearn.model_selection import GridSearchCV
```

```python
classifier = GridSearchCV(SVC(gamma='auto'), {
	'C': [1, 10, 20, 30, 40],
	'kernel': ['rbf', 'linear', 'sigmoid', 'poly']
}, cv=5, return_train_score=False)
classifier.fit(df.drop(['flower'], axis='columns'), df.flower)
```

```python
classifier.cv_results_
```

```python
result_df = pd.DataFrame(classifier.cv_results_)
result_df
```

```python
result_df[['param_C', 'param_kernel', 'mean_test_score']]
```

```python
classifier.best_score_
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
model_param ={
	'svm': {
		'model': SVC(),
		'param': {
			'kernel': ['rbf', 'linear', 'sigmoid', 'poly'],
			'C': [1, 10, 20, 30, 40]
		}
	},
	'random_forest': {
		'model': RandomForestClassifier(),
		'param': {
			'n_estimators': [1, 5, 10]
		}
	},
	'logistic_regression': {
		'model': LogisticRegression(solver='lbfgs', max_iter=1000),
		'param': {
			'C': [1, 5, 10]
		}
	}

}
```

```python
score = []

for model_name, mp in model_param.items():
	clf = GridSearchCV(mp['model'], mp['param'], cv=5, return_train_score=False)
	clf.fit(df.drop(['flower'], axis='columns'), df.flower)
	score.append({
		'model': model_name,
		'best_score': clf.best_score_,
		'best_params': clf.best_params_
	})

df = pd.DataFrame(score, columns=['model', 'best_score', 'best_params'])
df
```
