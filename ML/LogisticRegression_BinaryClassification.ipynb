```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
age_df = pd.read_csv('data/buy_insurance.csv')
age_df.head()
```

```python
plt.xlabel('Age')
plt.ylabel('Bought Insurance')
plt.scatter(age_df.age, age_df.bought_insurance, color='red', marker='+')
plt.show()
```

```python
X_train, X_text, y_train, y_test = train_test_split(age_df[['age']], age_df.bought_insurance, test_size=0.3)
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
model.predict(X_text)
```

```python
y_test
```

```python
model.score(X_text, y_test)
```

```python
model.predict_proba(X_text)
```
