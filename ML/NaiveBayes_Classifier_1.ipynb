```python

import pandas as pd
```

```python
titanic_df = pd.read_csv('data/titanic.csv')
titanic_df.head()
```

```python
titanic_df.drop(["PassengerId", "Name", "SibSp", "Ticket", "Cabin", "Embarked"], axis=1, inplace=True)
titanic_df.head()
```

```python
target = titanic_df["Survived"]
features = titanic_df.drop("Survived", axis=1)
```

```python
dummies = pd.get_dummies(features.Sex)
features = pd.concat([features, dummies], axis=1)
features.drop('Sex', axis=1, inplace=True)
features.head()
```

```python
features.columns[features.isnull().any()]
```

```python
input_features = features.fillna(features.Age.mean(), inplace=True)
```

```python
features.columns[features.isnull().any()]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import cross_val_score
```

```python
model = GaussianNB()
```

```python
cross_val_score(model, X_train, y_train, cv=5)
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
model.predict_proba(X_test)
```
